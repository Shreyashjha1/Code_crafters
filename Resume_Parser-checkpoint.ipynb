```python
import fitz
import docx
import re
```

#Read Resumes

```python
def read_pdf(path):
    text=""
    pdf=fitz.open(path)
    for page in pdf:
        text += page.get_text()
    return path
def read_docx(path):
    doc = docx.Document(path)
    return "\n".join(p.text for p in doc.paragraphs)
```

#Clean Resume Text

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

#Skills Dictonary

```python
SKILLS = [
    "python", "java", "sql", "machine learning",
    "data science", "deep learning",
    "flask", "django", "excel", "react"
]
```

#Extract Skills

```python
def extract_skills(text):
    return [skill for skill in SKILLS if skill in text]
```

#Extract Education

```python
def extract_education(text):
    for edu in ["b.tech", "bachelor", "master", "m.tech", "msc", "phd"]:
        if edu in text:
            return edu
    return "Not Found"
```

#Extract Experience

```python
def extract_experience(text):
    match = re.search(r'(\d+)\s+years?', text)
    return match.group(1) + " years" if match else "Fresher / Not Mentioned"
```

#Combining Every Logics

```python
def parse_resume(file_path):
    if file_path.endswith(".pdf"):
        raw = read_pdf(file_path)
    else:
        raw = read_docx(file_path)

    clean = clean_text(raw)

    return {
        "skills": extract_skills(clean),
        "education": extract_education(clean),
        "experience": extract_experience(clean)
    }
```
